```python
import tweepy 
from textblob import TextBlob
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import re 
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
dataset = pd.read_csv('dataset.csv')
dataset
```

```python
def con(x):
    if(x==2):
        return 'Neutral'
    elif (x==4):
        return 'Positive'
    else :
        return 'Negative'
       
dataset['analysis']=dataset['sentiment'].apply(con)
dataset
```

```python
plt.title('Sentiment Analysis')
dataset['analysis'].value_counts().plot(kind = 'pie',figsize=(5, 5))
plt.savefig('plot.svg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
dataset
```

```python
#word cloud
allWords = ' '.join([twts for twts in dataset['tweet']])

wordCloud = WordCloud(max_font_size=500, max_words=100, background_color="white").generate(allWords)

plt.imshow(wordCloud, interpolation="bilinear")
plt.axis('off')
plt.show()

wordCloud.to_file('Image.png')
print('Image Saved Sucessfullly')
```
