```python
import tweepy 
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import re 
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
dataset = pd.read_csv('dataset.csv')
dataset
```

```python
dataset
```

```python
#word cloud
allWords = ' '.join([twts for twts in dataset['tweet']])

wordCloud = WordCloud(max_font_size=500, max_words=200, background_color="white").generate(allWords)

plt.imshow(wordCloud, interpolation="bilinear")
plt.axis('off')
plt.show()

wordCloud.to_file('Image.png')
print('Image Saved Sucessfullly')
```
